```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit_transform(df)
X_scaler
```

```python
pca = PCA(n_components = 0.9, random_state = 42)
X_pca = pca.fit_transform(X_scaler)
```

```python
explained_variance = pca.explained_variance_ratio_
print("explained_variance", explained_variance)
```

```python
cumulative_variance = np.cumsum(explained_variance)
print("cumulative_variance", cumulative_variance)
```

```python
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker = 'o', color = "b")
plt.title("Cumulative explained variance by PCA components")
plt.xlabel("Numbers of principal components Analysis")
plt.ylabel("Cumulative Explained variance")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns#
sns.scatterplot(x = X_pca[:, 0], y = X_pca[:, 1])
plt.title("principal component analysis")
plt.xlabel("principal components (pca)1")
plt.ylabel("principal component (pca)2 ")
plt.legend()
plt.show()
```

```python
kmean = KMeans(n_clusters = 4, random_state = 42)
kmean.fit(X_scaler)
```

```python
y_pred = kmean.predict(X_scaler)
```

```python
score = silhouette_score(X_scaler, y_pred)
score
```

```python
wcss = []
for i in range(2, 21):
    km = KMeans(n_clusters = i, random_state = 42, n_init = "auto")
    km.fit(X_scaler)
    wcss.append(km.inertia_)

plt.plot(range(2, 21), wcss, color = "Blue", marker = "o")
plt.title("elbows method")
plt.xlabel("Number of cluster")
plt.ylabel("inertia")
plt.show()
    
```

```python
km = KMeans(n_clusters = i, random_state = 42, n_init = "auto")
km.fit(X_scaler)
c = km.labels_
```

```python
score = silhouette_score(X_scaler, c, metric = "euclidean")
score
```

```python
from sklearn.metrics import davies_bouldin_score
```

```python
scored = davies_bouldin_score(X_scaler, c)
scored
```
