```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import root_mean_squared_error as mse
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree  import DecisionTreeRegressor
from sklearn.feature_extraction import DictVectorizer
```

```python
house = pd.read_csv("housing.csv")
house
```

```python
str = list(house.dtypes[house.dtypes == "object"].index)
str
```

```python
for col in str:
    house[col] = house[col].str.lower()
    
```

```python
house = house[house["ocean_proximity"].isin(["<ihocean", "inland"])]
house
```

```python
house["ocean_proximity"]
```

```python
house = house.fillna(0)
house
```

```python
seed = np.random.seed(1)
seed
```

```python
df_full_train, df_test = train_test_split(house, test_size =0.2, random_state = seed)
df_train, df_val = train_test_split(df_full_train, test_size = 0.25, random_state = seed)
```

```python
len(df_train), len(df_val), len(df_test)
```

```python
df_train = df_train.reset_index(drop = True)
df_val = df_val.reset_index(drop = True)
df_test = df_test.reset_index(drop = True)
```

```python
y_train = np.log1p(df_train.median_house_value.values)
y_val = np.log1p(df_val.median_house_value.values)
y_test = np.log1p(df_test.median_house_value.values)
```

```python
del df_train["median_house_value"]
del df_val["median_house_value"]
del df_test["median_house_value"]
```

```python
train_dict = df_train.fillna(0).to_dict(orient = "records")
dv = DictVectorizer(sparse =False)
X_train = dv.fit_transform(train_dict)
X_train
```

```python
dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
```

```python
val_dict = df_val.fillna(0).to_dict(orient = "records")
X_val = dv.transform(val_dict)
```

```python
dt = DecisionTreeRegressor(max_depth = 1)
dt.fit(X_train, y_train)
```

```python
print(export_text(dt, feature_names = list(dv.get_feature_names_out())))
```

```python
rf = RandomForestRegressor(n_estimators = 10, random_state = seed, n_jobs = 1)
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_val)
rmse = mse(y_val, y_pred)
rmse
```

```python
scores = []
n_estimators = [10, 201, 10]

for n in n_estimators:
    rf = RandomForestRegressor(n_estimators = n, random_state = 1, n_jobs = -1)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_val)
    rmse = mse(y_val, y_pred)
    scores.append((rmse, n))
```

```python
print(n, rmse)
```

```python
scores = []
n_estimators = [10, 201, 10]
for max_depth in [10, 15, 20, 25]:
    for n in n_estimators:
        
       rf = RandomForestRegressor(n_estimators = n, max_depth = max_depth, random_state = 1, n_jobs = -1)
       rf.fit(X_train, y_train)
       y_pred = rf.predict(X_val)
       rmse = mse(y_val, y_pred)
       scores.append((rmse, n_estimators, max_depth))
```

```python
print(n, rmse, max_depth)
```

```python
rf = RandomForestRegressor(n_estimators = 10, max_depth = 20, random_state =1, n_jobs = -1)
rf.fit(X_train, y_train)
```

```python
importance_features = pd.Series(rf.feature_importances_, index = dv.get_feature_names_out())
importance = importance_features.sort_values(ascending = False)

plt.figure(figsize = (5,6))
importance.plot(kind = "bar")
plt.title("features of using random forest Regressor")
plt.xlabel("features")
plt.ylabel("importance")
plt.tight_layout()
plt.show()
```
