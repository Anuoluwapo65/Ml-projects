```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn.feature_extraction import DictVectorizer
```

```python
exam = pd.read_csv("jamb_exam_results.csv")
exam.head()
```

```python
exam.columns = exam.columns.str.lower().str.replace(' ', '_')
exam.columns
```

```python
strings = list(exam.dtypes[exam.dtypes == "object"].index)
strings
```

```python
for col in strings:
    exam[col] = exam[col].str.lower()
    exam
```

```python
exam.head()
```

```python
del exam["student_id"]
exam
```

```python
exam.isnull().sum()
```

```python
exam = exam.fillna(0)
exam.head()
```

```python
exam.isnull().sum()
```

```python
seed = np.random.seed(1)
df_full_train, df_test = train_test_split(exam, test_size = 0.2, random_state = seed)
df_train, df_val = train_test_split(df_full_train, test_size = 0.25, random_state = seed)
```

```python
df_train = df_train.reset_index(drop = True)
df_val = df_val.reset_index(drop = True)
df_test = df_test.reset_index(drop = True)
```

```python
y_train  = df_train.jamb_score.values
y_val = df_val.jamb_score.values
y_test = df_test.jamb_score.values
```

```python
del df_train["jamb_score"]
del df_val["jamb_score"]
del df_test["jamb_score"]
```

```python
train_dict = df_train.fillna(0).to_dict(orient = "records")
dv = DictVectorizer(sparse = False)
X_train = dv.fit_transform(train_dict)
X_train
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
val_dict = df_val.fillna(0).to_dict(orient = "records")
X_val = dv.transform(val_dict)
```

```python
dt = DecisionTreeClassifier(max_depth = 1)
dt.fit(X_train, y_train)
```

```python
from sklearn.tree import export_text
print(export_text(dt, feature_names = list(dv.get_feature_names_out())))
```

```python

rf = RandomForestRegressor(n_estimators= 10, random_state = seed, n_jobs = 1)
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import root_mean_squared_error as mse
```

```python
y_pred = rf.predict(X_val)
rmse = mse(y_val, y_pred)
rmse
```

```python
scores = []

for n in range(10, 201, 10):
    rf = RandomForestRegressor(n_estimators = n, random_state = seed, n_jobs = -1)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_val)
    rmse = mse(y_val, y_pred)
    scores.append((rmse, n))
```

```python
print(rmse, n)
```

```python
print('%4s --> %.3f' % (n, rmse))
```

```python
scores = []
for n in range(10, 201, 10):
    for t in [10, 15, 20, 25]:
        rf = RandomForestRegressor(n_estimators = n, random_state = 1, max_depth = t, n_jobs = 1)
        rf.fit(X_train, y_train)
        y_pred = rf.predict(X_val)
        rmse = mse(y_val, y_pred)
        scores.append((n, t, rmse))
```

```python
print(n, t, rmse)
```

```python
rf = RandomForestRegressor(n_estimators = 10, max_depth = 20, random_state =1, n_jobs = -1)
rf.fit(X_train, y_train)
```

```python
importance_features = pd.Series(rf.feature_importances_, index = dv.get_feature_names_out())
importance = importance_features.sort_values(ascending = False)

plt.figure(figsize = (5,6))
importance.plot(kind = "bar")
plt.title("features of using random forest Regressor")
plt.xlabel("features")
plt.ylabel("importance")
plt.tight_layout()
plt.show()
```

```python
import xgboost as xgb
```

```python
features = list(dv.get_feature_names_out())
dtrain = xgb.DMatrix(X_train, label = y_train, feature_names = features)
dval = xgb.DMatrix(X_val, label = y_val, feature_names = features)
watchlist = [(dtrain, "train"), (dval, "val")]
```

```python
xgb_params = {
    "eta":0.3,
    "max_depth":6,
    "min_child_length":1,

    "objective":"reg:squarederror",
    "nthread":8,

    "seed":1,
    "verbosity":1
}


model = xgb.train(xgb_params, dtrain, num_boost_round = 100, verbose_eval = 5, evals = watchlist)
```
