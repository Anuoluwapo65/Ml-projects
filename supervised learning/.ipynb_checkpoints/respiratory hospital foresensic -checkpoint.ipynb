```python
import seaborn as sns
```

```python
data["week ending date"] = pd.to_datetime(data["week ending date"])

data["absolute change in the percent hospitals reporting percent inpatient beds occupied by covid-19 patients from prior week"]*=100
fig = plt.figure(figsize=(10,10))
sns.lineplot(data = data, x = "week ending date", y =  "absolute change in the percent hospitals reporting percent inpatient beds occupied by covid-19 patients from prior week",color = "r", marker = "o")
```

```python
data["number of adult icu beds"] = pd.to_datetime(data["number of adult icu beds"])
data["absolute change in the percent hospitals reporting rsv admissions from prior week"] *= 100

fig = plt.figure(figsize=(10,10))
sns.lineplot(data = data, x = "number of adult icu beds", y = "absolute change in the percent hospitals reporting rsv admissions from prior week" , color = "y", marker = "o")
```
